```python
#Histograms
#Made up of vertical bars that track frequency of data 

import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
body_data = pd.read_csv("weights_and_heights.csv")
```

```python
#Split data frames

male = body_data[body_data["Gender"] == "Male"]
female = body_data[body_data["Gender"] == "Female"]
```

```python
#Historgram for males
plt.style.use("seaborn-poster")
plt.hist(male["Weight"])
plt.show()
```

```python
#We can split into further bins
#Bins by default is set by 10s
#More bins can demonstrate whether data is evenly distributed
#Historgram for males
plt.style.use("seaborn-poster")
plt.hist(male["Weight"], bins = 20, edgecolor = "black")
plt.show()
```

```python
#Males and females
plt.style.use("seaborn-poster")
plt.hist(male["Weight"], bins = 20, edgecolor = "black")
plt.hist(female["Weight"], bins = 20, edgecolor = "black")
plt.show()
```

```python
#It can be difficult to see where the overlap is here
#Female plotted second - plotted right on top of male
#We can fix this with alpha parameter - allows control over transparency over plot = 1 - full color, toward 0 more transparent
#Males and females
plt.style.use("seaborn-poster")
plt.hist(male["Weight"], bins = 20, edgecolor = "black", alpha = 0.6)
plt.hist(female["Weight"], bins = 20, edgecolor = "black", alpha = 0.6)
plt.show()
```

```python
#Males and females
plt.style.use("seaborn-poster")
plt.hist(male["Weight"], bins = 20, edgecolor = "black", alpha = 0.6, color = "royalblue", label = "Male")
plt.hist(female["Weight"], bins = 20, edgecolor = "black", alpha = 0.6, color = "magenta", label = "Female")
plt.title("Distribution of Body Weight by Gender")
plt.xlabel("Weight (lbs)")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```
